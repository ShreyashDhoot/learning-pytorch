### Matrix multiplication
2 main ways to do matrix multiplication in ML and DL
1. element wise matric multiplication
2. matrix multiplication (dot product)

```python
## element wise
tensor_a=torch.rand(3,3)
tensor_b=torch.rand(3,3)
print(tensor_a)
print(tensor_b)
print(f"proof look at left corner element of both the matrix \n {tensor_a[0][0]:.4g}*{tensor_b[0][0]:.4g}={tensor_a[0][0] * tensor_b[0][0]:.4g}")
print(tensor_a*tensor_b)
```

```python
#matrix multiplication
tensor_1=torch.rand(3,4)
tensor_2=torch.rand(4,3)
## matmul is depricated so use torch.mm
new_tensor=torch.mm(tensor_1,tensor_2)
print(new_tensor.size())
print(new_tensor)
```

```python
#shapes for matrix multiplication
tensor_A=torch.tensor([[1,2],
                       [3,4],
                       [5,6]])
tensor_B=torch.tensor([[7,10],
                       [8,11],
                       [9,12]])
##using a transpose we can multiply these
print(f"Tensor B-> {tensor_B.shape} \n {tensor_B}")
print(f"transpose of Tensor B-> {tensor_B.T.shape} \n {tensor_B.T}")
torch.mm(tensor_A,tensor_B.T)
```

### Reshaping , stacking , Squeezing and unsqueezing
1. Reshaping - reshpaes an input to a definite shape
2. view - return the view of an input tensor of certain shape but keep ythe same memory as the original tensor
3. Stacking- combining multiple tensors on each other (v.stack-> vertical stack ) (h.stack-> horizontal stack )
4. Squeeze - removes all '1' dimensions from a tensor
5. Unsqueeze - add a '1' dimension to the tensor
6. Permute - return a view of the input with the dimension permuted (swapped) in a certain way

```python
x=torch.arange(1,10)
x,x.shape
```

```python
# add an extra dimension
# reshape has to be compatible with the number of elements
x_reshape = x.reshape(1,9)
x_reshape,x_reshape.shape
```

```python
# change the view
# z is just a different view of x
# changing view of z changes x
z=x.view(1,9)
z,z.shape
```

```python
# stack tensors on top of each other
x_stacked = torch.stack([x,x,x,x],dim=0)
x_stacked,x_stacked.shape
```

```python
#squeeze
x_squeezed=torch.squeeze(x_stacked,dim=0)
x_squeezed,x_squeezed.shape
```

```python
#unsqueeze
x_unsqueeze=torch.unsqueeze(x_squeezed,dim=0)
x_unsqueeze, x_unsqueeze.size()
```

```python
#permute
torch.permute(x_unsqueeze,(0,1,2))
```

```python
torch.permute(x_unsqueeze,(1,0,2))
```

```python
x_original=torch.rand(size=(224,224,3))
# permute the original tensor to rearrage the axis (or dim) order
x_permuted=x_original.permute(2,0,1)
print("previous shape :\n")
print(x_original.size())
print("permuted shape :\n")
print(x_permuted.size())
```
